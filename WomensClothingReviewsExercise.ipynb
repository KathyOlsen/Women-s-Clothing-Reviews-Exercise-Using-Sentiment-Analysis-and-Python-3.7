```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer  
from string import punctuation
from textblob import TextBlob
```

```python
Location = "Womens Clothing E-Commerce Reviews.csv"
df = pd.read_csv(Location)
df.head()
```

```python
len(df)
```

```python
df.columns = ['Unnamed: 0','Clothing ID','Age','Title','Review Text','Rating','Recommended IND','Positive Feedback Count','Division Name','Department Name','Class Name']
df = df.drop('Unnamed: 0', axis=1)
df1 = df.dropna(axis=0,subset=['Review Text'])
df1 = df1.drop_duplicates()
for review in df1['Review Text']:
    dtype = str

df1.head()
```

```python
len(df1)
```

```python
def reviewSentiment(review):
    
    review = review.lower()
    
    tknz_review = word_tokenize(review)
    
    for token in tknz_review:
        if token in punctuation:
            tknz_review.remove(token)
    
    clean_tokens = []
    eng_stopwords = stopwords.words('english')
    for token in tknz_review:
        if token not in eng_stopwords:
            clean_tokens.append(token)
            
    clean_review = ' '.join(clean_tokens)
    
    blob_rev = TextBlob(clean_review)
    
    sent_pol = blob_rev.sentiment.polarity
    
    return sent_pol
```

```python
df1['Review Sentiment'] = df1['Review Text'].apply(reviewSentiment)
df1.head()
```

```python
def sentimentCategory(sent_num):
    if sent_num >= 0.2:
        return "positive"
    if sent_num <= -0.2:
        return "negative"
    else:
        return "neutral"
```

```python
df1['Sentiment Category'] = df1['Review Sentiment'].apply(sentimentCategory)
df1.head()
```
